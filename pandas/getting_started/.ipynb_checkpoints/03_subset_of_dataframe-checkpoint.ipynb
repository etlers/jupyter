```python
import pandas as pd
```

```python
titanic = pd.read_csv("data/titanic.csv")
titanic.head()
```

```python
# Each column in a DataFrame is a Series. As a single column is selected, the returned object is a pandas Series.
ages = titanic["Age"]
ages.head()
```

```python
# We can verify this by checking the type of the output.
type(titanic["Age"])
```

```python
# And have a look at the shape of the output.
# DataFrame.shape is an attribute of a pandas Series and DataFrame containing the number of rows and columns: (nrows, ncolumns)
# A pandas Series is 1-dimensional and only the number of rows is returned
titanic["Age"].shape
```

```python
age_sex = titanic[["Age", "Sex"]]
age_sex.head()
```

```python
# The returned data type is a pandas DataFrame
type(titanic[["Age", "Sex"]])
```

```python
# A DataFrame is 2-dimensional with a row and column dimension 
titanic[["Age", "Sex"]].shape
```

```python
above_35 = titanic[titanic["Age"] > 35]
above_35.head()
```

```python
titanic["Age"] > 35
```

```python
above_35.shape
```

```python
class_23 = titanic[titanic["Pclass"].isin([2,3])]
class_23.head()
```

```python
# When combining multiple contional statements, each condition must be serrounded by parentheses ().
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23.head()
```

```python
age_no_na = titanic[titanic["Age"].notna()]
age_no_na.head()
```

```python
age_no_na.shape
```

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names.head()
```

```python
adult = titanic.loc[titanic["Age"] > 35]
adult.head()
```

```python
# rows 10 till 25 and columns 3 to 5
titanic.iloc[9:25, 2:5]
```

```python
# When selecting specific rows and/or columns with loc or iloc, new values can be assigned to the selected data.
titanic.iloc[0:3, 3] = "anonymous"
titanic.head()
```
