```python
def draw_jongmok_plot(nm):
    plt.rc("font", family = "AppleGothic")
    sns.set(
        font="AppleGothic", 
        rc={"axes.unicode_minus":False}, style='darkgrid'
    )

    if nm == "KODEX 레버리지":
        df_jongmok_plot = df_kodex_diff
    else:
        df_jongmok_plot = df_deal_source[df_deal_source.JONGMOK_NM == nm]
    p = df_jongmok_plot.plot(x="TIMES", y=["CPRICE"])
    p.set_title(nm)
```

```python
draw_jongmok_plot("LG에너지솔루션")
```

```python
list_deal_jongmok = df_deal_source["JONGMOK_NM"].unique()

draw_jongmok_plot("KODEX 레버리지")

for nm in list_deal_jongmok:
    draw_jongmok_plot(nm)
```

```python
dict_tick = {
    "KODEX 레버리지": 5,
    "삼성전자": 100,
    "NAVER": 1000,
}

def apply_diff(df, nm):
    df["DIFF"] = df["DIFF"].apply(lambda x: x / dict_tick[nm] if x >= 0 else (x / dict_tick["KODEX 레버리지"]) * -1)
    
    return df

df_selected = df_deal_history[
    (df_deal_history["JONGMOK_NM"] == "삼성전자") | (df_deal_history["JONGMOK_NM"] == "NAVER")
][["JONGMOK_NM", "TIMES", "DIFF"]]

df_kodex = df_deal_history[df_deal_history["JONGMOK_NM"] == "KODEX 레버리지"]
df_kodex = apply_diff(df_kodex, "KODEX 레버리지")[["JONGMOK_NM", "TIMES", "DIFF"]]

df_result = pd.concat([df_selected, df_kodex])
df_result
```

```python
sns.relplot(x="TIMES", y="DIFF", hue="JONGMOK_NM", 
            aspect=2,
            kind="line", 
            data=df_result)
```
