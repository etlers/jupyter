```python
import pymysql
import pandas as pd
```

```python
def get_qry_data_list(qry):
    # MySQL Connection 연결
    conn = pymysql.connect(
        host='localhost', 
        user='etlers', 
        password='wndyd',
        db='stocks', 
        charset='utf8')

    # Connection 으로부터 Cursor 생성
    curs = conn.cursor()
    # SQL문 실행
    curs.execute(qry)
    # 데이타 Fetch
    fetched_list = curs.fetchall()
    # Connection 닫기
    conn.close()
    
    return fetched_list
```

```python
def show_jongmok_sum(df_base, jongmok_cd):
    inc_deal_tf = True
    inc_forgn_tf = True
    pre_day_forgn = 0
    df_filtered = df_base.loc[df_base.jongmok_cd == jongmok_cd]
    df_filtered = df_filtered.sort_values("deal_dt", axis=0, ascending=False).head(5)
    # 기관 거래량이 늘고 외국인 보유가 늘어났다면
    for idx, row in df_filtered.iterrows():
        if pre_day_forgn > 0:
            if row["forgn_amount"] <= pre_day_forgn:
                inc_forgn_tf = False
            else:
                inc_forgn_tf = True
    pre_day_forgn = row["forgn_amount"]
    
    if inc_forgn_tf == True:
        print(df_filtered)
```

```python
def get_jongmok_sum():
    list_cols = [
        "jongmok_cd",
        "jongmok_nm",
        "deal_dt",
        "buy_amount",
        "sell_amount",
        "forgn_amount",
    ]
    # SQL문 실행
    sql = """
    select jongmok_cd 
         , jongmok_nm
         , deal_dt
         , buy_amount
         , sell_amount
         , forgn_amount
      from jongmok_sum
    """
    list_jongmok_sum = get_qry_data_list(sql)    
    df_jongmok_sum = pd.DataFrame(list_jongmok_sum, columns=list_cols)
    df_jongmok_cd = df_jongmok_sum[["jongmok_cd"]].drop_duplicates()
    
    for idx, row in df_jongmok_cd.iterrows():
        show_jongmok_sum(df_jongmok_sum, row["jongmok_cd"])
```
